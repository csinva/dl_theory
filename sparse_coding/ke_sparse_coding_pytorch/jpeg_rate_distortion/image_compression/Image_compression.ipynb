```python
import sigpy.plot as pl
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
# os.system("ls -l")
%matplotlib notebook
```

```python
def rmse(im_g,im_r):
    return np.sqrt(np.sum(((im_g-im_r)**2))/np.size(im_g))
```

```python
rmses = []
bits = []
groundtruth = cv2.imread("lena_gray_512.tif",cv2.IMREAD_GRAYSCALE)
for i in range(31):
    dir1 = "image_output%s.jpg" %str(i+1)
    result = cv2.imread(dir1,cv2.IMREAD_GRAYSCALE)
    rmse_1 = rmse(result,groundtruth)
    bit = (os.path.getsize(dir1)*8)/(512*512)
    print(bit)
    rmses.append(rmse_1)
    bits.append(bit)
```

```python
plt.plot(rmses,bits);plt.title("Rate Distortion Performance")
plt.xlabel("Distortion")
plt.ylabel("Rate")
plt.show()
```
